Cargo los datos, tomo solo las 10mil palabras mas frecuentes del dataset de todas las noticias juntas. Cada palabra tendrá asignado un número
 

```python
from tensorflow.keras.datasets import reuters
(train_data, train_labels), (test_data, test_labels) = reuters.load_data(
    num_words=10000)
```

Datos de entrada a la red neuronal. Alimentaremos la red neuronal con 8982 vectores (uno por noticia) de 10000 componentes cada uno. Entonces para cada noticia, pondremos '1' en el word_index de cada palabra que aparece en el mensaje y ceros en los demás lugares.  

```python
import numpy as np
def vectorize_sequences(sequences, dimension=10000):
  results = np.zeros((len(sequences), dimension))
  for i, sequence in enumerate(sequences):
    results[i, sequence] = 1.
  return results
```

```python
x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)
```

Puedo corroborar que los vectores de entrada de la red son distintos, por ejemplo sumando los valores 1 para cada componente de x_train y viendo que dan distinto.

```python
print(sum(x_train[5]),sum(x_train[10]),sum(x_train[25]))
```

A la salida de mi red neuronal, tendré 8982 vectores de 46 componentes cada uno. En el vector de salida de cada noticia habrá un 1 en la categoría de esa noticia y 0 en las demás categorías. Es importante tener en cuenta que en este caso, una dada noticia debe clasificarse en solo una de las 46 categorías.

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(train_labels)
y_test = to_categorical(test_labels)
```

```python
print(sum(sum(y_train)))
```

Armo la red neuronal que tendrá como entrada un vector 8982 componentes (len de train_data) y cada una de esas componentes será a su vez un vector de 100000 componentes.

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(64, activation="relu"),
    layers.Dense(64, activation="relu"),
    layers.Dense(46, activation="softmax")
])
```

Me armé una red neuronal con dos capas internas de 64 neuronas cada una y una capa final de salida de 46 neuronas que serán las categorías de clasificación de las noticias. La capa de salida será comparada con los labels verdaderos usando una función de costo. La mejor función de costo para usar en este caso es la de "categorical_crossentropy". Esta función mide la distancia entre dos distribuciones de probabilidad, la de la salida de la red y la de los labels verdaderos.Minimizando la distacia entre ambas distribuciones, entrenamos la red para que los vectores de slida se parezcan lo más posible a las etiquetas verdaderas.

```python
model.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])
```

**Validación**

Separo una parte de los datos de entrenamiento para realizar la validación

```python
x_val = x_train[:1000]
partial_x_train = x_train[1000:]
y_val = y_train[:1000]
partial_y_train = y_train[1000:]
```

Entreno con los datos parciales

```python
history = model.fit(partial_x_train,
                    partial_y_train,
                    epochs=20,
                    batch_size=512,
                    validation_data=(x_val, y_val))
```

```python
history_dict = history.history
history_dict.keys()
```

Grafico la función de costo para el set de test y el de validación.

```python
import matplotlib.pyplot as plt
loss = history.history["loss"]
val_loss = history.history["val_loss"]
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, "bo", label="Training loss")
plt.plot(epochs, val_loss, "b", label="Validation loss")
plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

Reentreno el modelo para 9 Epochs porque veo del gráfico anterior que a partir de de ese número el modelo comienza a overfitear.

```python
model = keras.Sequential([
  layers.Dense(64, activation="relu"),
  layers.Dense(64, activation="relu"),
  layers.Dense(46, activation="softmax")
])
model.compile(optimizer="rmsprop",
              loss="categorical_crossentropy",
              metrics=["accuracy"])
model.fit(x_train,
          y_train,
          epochs=9,
          batch_size=512)
results = model.evaluate(x_test, y_test)
```

```python
import copy
test_labels_copy = copy.copy(test_labels)
np.random.shuffle(test_labels_copy)
hits_array = np.array(test_labels) == np.array(test_labels_copy)
hits_array.mean()
```

Este resultado muestra que comparada con una clasificación random, la precisión alcanzada por nuestra red es del 80%%.
